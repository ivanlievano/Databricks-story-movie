```sql
describe extended demo.results_movie_genre_language_py;
```

```sql
create table if not exists demo.results_movie_genre_language_sql(
  title	string,
  duration_time	string,
  release_date	string,
  vote_average	string,
  language_name	string,
  genre_name	string,
  created_date	timestamp
)

using parquet
location '/mnt/storymovie/gold/results_movie_genre_language_external_sql'
```

```sql
insert into demo.results_movie_genre_language_sql
select * 
from demo.results_movie_genre_language_py
where
    genre_name = 'Adventure'
```

#Vistas

```sql
create or replace temp view v_results
    as
      select * 
      from demo.results_movie_genre_language_py
          where
              genre_name = "Adventure"
```

```sql
select *
from v_results;
```

#Crear vistas globales

```sql
create or replace global temp view v_results
    as
      select * 
      from demo.results_movie_genre_language_py
          where
              genre_name = "Drama"
```

```sql
select * 
from global_temp.v_results
          
```

#Crear vistas permanentes

```sql
create or replace view v_results_permanent
    as
      select * 
      from demo.results_movie_genre_language_py
          where
              genre_name = "Comedy"
```

```sql
select * from v_results_permanent
```
