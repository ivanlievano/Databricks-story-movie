```python
from pyspark.sql.functions import col, current_timestamp, lit
```

```python
genre_df = spark.read.option("header", "True") \
                    .option("inferSchema", "True") \
                    .csv("/mnt/storymovie/bronze/genre.csv")
```

```python
genre_renamed_df = genre_df.withColumnRenamed("genreId", "genre_id") \
                            .withColumnRenamed("genreName", "genre_name")
```

```python
genre_final_df = genre_renamed_df.withColumn("ingestion_date", current_timestamp()) \
                                .withColumn("env", lit("python"))
```

```python
genre_final_df.write.mode("overwrite").parquet("/mnt/storymovie/silver/genres")
```
