```python
from pyspark.sql.functions import current_timestamp, lit, concat, col
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
```

```python
production_company_schema = StructType(fields=[
    StructField("company_id", IntegerType(), True),
    StructField("company_name", StringType(), True)
])
```

```python
production_company_df = spark.read \
                .schema(production_company_schema) \
                .csv('/mnt/storymovie/bronze/production_company')
```

```python
production_company_final_df = production_company_df \
                            .withColumn("ingestion_date", current_timestamp()) \
                            .withColumn("env", lit("python"))
```

```python
production_company_final_df.write.mode("overwrite").parquet('/mnt/storymovie/silver/productions_companies')
```
