#Ingestion del archivo country.json

```python
from pyspark.sql.functions import current_timestamp, lit
```

## Paso 1: Leer archivo

```python
countries_schema = "countryId INT, countryIsoCode STRING, countryName STRING"
```

```python
countries_df = spark.read.option("header", "True") \
                    .schema(countries_schema) \
                    .json("/mnt/storymovie/bronze/country.json")
```

## Paso 2: Eliminar columna no deseada

```python
countries_dropped_df = countries_df.drop("countryIsoCode")
```

## Paso 3: Configuracion de columnas

```python
countries_renamed_df = countries_dropped_df.withColumnRenamed("countryId" , "country_id") \
                                    .withColumnRenamed("countryName", "country_name")
```

```python
countries_final = countries_renamed_df.withColumn('ingestion_date', current_timestamp()) \
                                        .withColumn('env', lit('python'))
```

## Paso 4: Escritura de tabla

```python
countries_final.write.mode("overwrite").parquet('/mnt/storymovie/silver/countries')
```
