```python
from pyspark.sql.functions import current_timestamp, lit, concat, col
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
```

```python
language_role_df = spark.read \
                .option("inferSchema" , "True") \
                .option("multiLine", "True") \
                .json('/mnt/storymovie/bronze/language_role.json')
```

```python
language_role_with_columns_df = language_role_df \
                            .withColumnRenamed("roleId", "role_id") \
                            .withColumnRenamed("languageRole", "language_role") \
                            .withColumn("ingestion_date", current_timestamp()) \
                            .withColumn("env", lit("python"))
```

```python
display(language_role_with_columns_df)
```

```python
language_role_with_columns_df.write.mode("overwrite").parquet('/mnt/storymovie/silver/languages_roles')
```
