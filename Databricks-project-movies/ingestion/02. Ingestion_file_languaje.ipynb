#Ingesta de archivo languaje.csv

```python
from pyspark.sql.functions import col, current_timestamp, lit
```

```python
languaje_df = spark.read \
    .option("header", "True") \
    .option("inferSchema", "True") \
    .csv("/mnt/storymovie/bronze/language.csv")
```

```python
languaje_select_df = languaje_df.select(col("languageId"), col("languageName"))
```

```python
language_renamed_df = languaje_select_df \
                    .withColumnRenamed("languageId", "language_id") \
                    .withColumnRenamed("languageName", "language_name")
```

```python
languages_final_df = language_renamed_df.withColumn("ingestion_date", current_timestamp()) \
                                        .withColumn("env", lit("python"))
```

```python
languages_final_df.write.mode("overwrite").parquet("/mnt/storymovie/silver/languages")
```
