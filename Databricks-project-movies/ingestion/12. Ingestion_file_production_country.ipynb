```python
from pyspark.sql.functions import current_timestamp, lit, concat, col
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
```

```python
schema = StructType(fields=[
    StructField("movieId", IntegerType(), True),
    StructField("countryId", IntegerType(), True)
])
```

```python
production_country_df = spark.read \
                .schema(schema) \
                .option("multiLine", "True") \
                .json('/mnt/storymovie/bronze/production_country')
```

```python
production_country_final_df = production_country_df \
                            .withColumnRenamed("movieId", "movie_id") \
                            .withColumnRenamed("countryId", "country_id") \
                            .withColumn("ingestion_date", current_timestamp()) \
                            .withColumn("env", lit("python"))
```

```python
production_country_final_df.write.mode("overwrite").parquet('/mnt/storymovie/silver/productions_countries')
```
