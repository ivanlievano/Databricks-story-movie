#Crear tablas para archivos csv
- Crear la tabla "movies"

```sql
drop table if exists movie_bronze.movies;
create table if not exists movie_bronze.movies(
  movieId int,
  tittle string,
  budget double,
  homePage string,
  overview string,
  popularity double,
  yearReleaseDate int,
  releaseDate date,
  revenue double,
  durationTime int,
  movieStatus string,
  tagline string,
  voteAverage double,
  voteCount int
)
using csv
options (path "/mnt/storymovie/bronze/movie.csv", header true)
```

```sql
show tables in movie_bronze
```

```sql
select * 
from movie_bronze.movies
```

```sql
drop table if exists movie_bronze.languajes;
create table if not exists movie_bronze.languajes(
  languageId int,
  languageCode string,
  languageName string
)

using csv
options(path "/mnt/storymovie/bronze/language.csv", header true)
```

```sql
select *
from movie_bronze.languajes
```

#Crear tablas para archivos JSON
- countries

```sql
drop table if exists movie_bronze.countries;
create table if not exists movie_bronze.countries(
  countryId int,
  countryIsoCode string,
  countryName string
)

using JSON
options (path "/mnt/storymovie/bronze/country.json", header true)
```

```sql
select *
from movie_bronze.countries
```

- tabla persons JSON multilinea

```sql
drop table if exists movie_bronze.persons;
create table if not exists movie_bronze.persons(
  personId int,
  personName struct<forename: string, surname: string>
)
using json
options (path "/mnt/storymovie/bronze/person.json", header true)
```

```sql
select *
from movie_bronze.persons
```

```sql
drop table if exists movie_bronze.movie_genres;
create table if not exists movie_bronze.movie_genres (
  movieId int,
  genreId int
)

using json
options(path "/mnt/storymovie/bronze/movie_genre.json", header true)
```

```sql
select *
from movie_bronze.movie_genres;
```

```sql
drop table if exists movie_bronze.movie_cast;
create table if not exists movie_bronze.movie_cast (
  movieId int,
  personId int,
  characterName String,
  genderId int,
  castOrder int
)

using json
options(path "/mnt/storymovie/bronze/movie_cast.json", multiLine true)
```
