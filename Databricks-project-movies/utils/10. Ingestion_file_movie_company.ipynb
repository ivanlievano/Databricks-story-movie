```python
from pyspark.sql.functions import current_timestamp, lit, concat, col
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
```

```python
%run "../includes/configuration"
```

```python
date = "2024-12-16"
```

```python
schema = StructType(fields=[
    StructField("movie_id", IntegerType(), True),
    StructField("company_id", IntegerType(), True)
])
```

```python
movie_company_df = spark.read \
                .schema(schema) \
                .csv(f'{bronze_folder_path}/{date}/movie_company')
```

```python
movie_company_final_df = movie_company_df \
                            .withColumn("ingestion_date", current_timestamp()) \
                            .withColumn("env", lit("python")) \
                            .withColumn("file_date", lit(date))
```

```python
movie_company_final_df.write.mode("overwrite").format("parquet").partitionBy('file_date').saveAsTable('movie_silver.movies_companies')
```
