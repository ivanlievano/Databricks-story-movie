```python
from pyspark.sql.functions import current_timestamp, lit, concat, col
```

```python
%run "../includes/configuration"
```

```python
date = "2024-12-30"
```

```python
movie_genre_df = spark.read \
                .option("inferSchema", "True") \
                .option("header", "True") \
                .json(f'{bronze_folder_path}/{date}/movie_genre.json')
```

```python
movie_genre_final = movie_genre_df \
                    .withColumnRenamed("genreId", "genre_id") \
                    .withColumnRenamed("movieId", "movie_id") \
                    .withColumn('ingestion_date', current_timestamp()) \
                    .withColumn('env', lit('python')) \
                    .withColumn("file_date", lit(date))
```

```python
movie_genre_final.write.mode('append').format("parquet").saveAsTable('movie_silver.movies_genres')
```
